```python
import pandas as pd
import matplotlib.pyplot as plt

results = pd.DataFrame({
    "Model": [
        "Logistic Regression",
        "LR + Momentum",
        "LR (7D)",
        "XGBoost",
        "RF Top 3",
        "RF + MACD",
        "RF + Returns",
        "RF + BB",
        "RF Baseline"
    ],
    "Accuracy": [
        49.22,
        48.87,
        50.66,
        50.66,
        51.01,
        51.25,
        51.49,
        51.61,
        52.09
    ]
})

plt.figure(figsize=(12,6))

bars = plt.bar(
    results["Model"],
    results["Accuracy"]
)

plt.axhline(
    52.09,
    linestyle="--",
    label="Best Model (RF Baseline)"
)

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height + 0.05,
        f"{height:.2f}",
        ha="center"
    )

plt.ylabel("Accuracy (%)")
plt.xlabel("Model / Experiment")
plt.title("Model and Feature Engineering Experiment Comparison")

plt.xticks(rotation=30)
plt.legend()

plt.ylim(48, 53)
plt.tight_layout()

plt.savefig(
    "../figures/model_comparison.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```
